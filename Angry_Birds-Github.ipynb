```python
'''
Author: Vighnesh Ghantasala 
Github: https://github.com/VighneshGhantasala/
'''

import cv2
import numpy as np
import pyautogui as pg
import time

cap = cv2.VideoCapture(0)
count  = 0
blue_obj = False 
m_down = False

while True:
    _, frame = cap.read()
    frame = cv2.flip(frame,1)
    #Rect for left click region
    cv2.rectangle(frame,(300,240),(640,480),(255,255,255))
    #Rect for Right click region
    cv2.rectangle(frame,(300,0),(640,240),(0,0,255))
    
    hsv = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)

    lower = np.array([0,153,11])   
    upper = np.array([12,255,255])
    #Light Blue [104,121,0] [137,255,255] #Dark blue [101,111,0] [125,251,255]
    lo_blue = np.array([101,111,0])
    up_blue = np.array([125,251,255])
    blue_mask = cv2.inRange(hsv,lo_blue,up_blue)
    mask = cv2.inRange(hsv,lower,upper)
    mask = cv2.dilate(mask,(5,5))
    result = cv2.bitwise_and(frame,frame,mask=mask)
    cont ,_ = cv2.findContours(mask,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    cont2 ,_ = cv2.findContours(blue_mask,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    for c in cont:
        area = cv2.contourArea(c)
        if area >= 3000 and area <= 5000:
            x , y ,w ,h = cv2.boundingRect(c)
            cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0))
            pg.moveTo(x,y)
            #Detect Blue object
            for cc in cont2:
                if cv2.contourArea(cc) >= 3000:
                    blue_obj = True
                    x1,y1,w1,h1 = cv2.boundingRect(cc)
                    cv2.rectangle(frame,(x1,y1),(x1+w1,y1+h1),(255,0,0))
                    if blue_obj == True:
                        time.sleep(0.1)
                        if x1 > 300 and y1 > 240:
#                             print("down")
                            pg.mouseDown( button='left')
                            m_down = True
                        if x1 > 300 and y1 < 240 and m_down == True:
#                             print("Up")
                            pg.mouseUp( button='left')
                            m_down = False
                        

                    


    cv2.imshow('frame',frame)    
    key = cv2.waitKey(1)
    if key == 27:
        break
        
cap.release()
cv2.destroyAllWindows()
```
